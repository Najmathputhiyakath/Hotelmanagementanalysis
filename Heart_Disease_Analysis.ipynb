**Split the columns into input and target variables (X and y)**

```python
x= df.drop(columns = ['target'])
y = df['target']
```

**Carry out Feature scaling using StandardScaler**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xcolumns = x.columns
```

```python
x = scaler.fit_transform(x)
```

```python
x= pd.DataFrame(x,columns = xcolumns)
```

```python
x.head()
```

**Split the dataset into training and testing set**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

**Machine Learning Models**

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
```

```python
abc = model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy =metrics.accuracy_score(y_pred,y_test)
accuracy
```

```python
metrics.confusion_matrix(y_pred,y_test)
```

```python
print(metrics.classification_report(y_pred,y_test))
```

```python
print (cross_val_score(abc, x, y, cv=10,scoring='accuracy').mean())
```

```python
sns.heatmap(metrics.confusion_matrix(y_pred,y_test),annot = True,fmt = 'd')
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'max_depth': [0.5, 1, 5, 10],
            'n_estimators': [16, 32, 50, 100], 'n_jobs': [1, 2],
            'random_state': [1, 10, 20, 42]}
```

```python
model = RandomForestClassifier()
```

```python
CV_rfc = GridSearchCV(estimator=model, param_grid=param_grid, cv= 5)
```

```python
CV_rfc.fit(x_train, y_train)
```

```python
print("Best parameters are:",CV_rfc.best_params_)
print('Best estimator is:',CV_rfc.best_estimator_)
print('Best score is:',CV_rfc.best_score_)
```

**Create Random forest model with the best parameters**

```python
rfc1=RandomForestClassifier(random_state=1, n_estimators= 50, max_depth=10, n_jobs=1)
```

```python
rfs=rfc1.fit(x_train,y_train)
```

```python
y_predict=rfc1.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
print("Accuracy score is",metrics.accuracy_score(y_test,y_predict))
print("Confusion metrics:",metrics.confusion_matrix(y_test,y_predict))
print("Classification report:",metrics.classification_report(y_test,y_predict))
print("Cross val score:",cross_val_score(rfs, x, y, cv=10,scoring='accuracy').mean())
```

**Visualize confusion matrix**

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_predict),annot=True,fmt = "d")
```
